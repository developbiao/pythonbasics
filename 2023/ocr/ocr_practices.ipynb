### Using Tesseract OCR

```python
import pytesseract
from PIL import Image
# Set the Tesseract executeable path
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

# Open the image using PIL
image = Image.open('114957.png')
print("Original Image")
display(image)

```

```python
# Preprocess the image (optional)
# Apply any necdessary preprocessing steps such as resizing, ddenosiing, or enhancing contrast

# Perform OCR
text = pytesseract.image_to_string(image, lang="chi_sim")

# Print the extracted test
print('------------ Recognize Result-------------')
print(text)
```

```python
### Using Concr package
from cnocr import CnOcr

def extract_text(result):
    lines = [''.join(line['text']) for line in result]
    text = '\n'.join(lines)
    return text

img_fg = '20230707115205.jpg'
ocr = CnOcr()
out = ocr.ocr(img_fg)
text = extract_text(out)
print(text)
```
