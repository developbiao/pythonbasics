```python
# Load image and preprocess it with vgg-16 structure
from tensorflow.keras.utils import load_img, img_to_array
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
import numpy as np
import os

model_vgg = VGG16(weights='imagenet', include_top=False)

# Define a methjod to load and preprocess the image
def modelProcess(img_path, model):
    img = load_img(img_path, target_size = (244, 244))
    img = img_to_array(img)
    x = np.expand_dims(img, axis = 0)
    x = preprocess_input(x)
    x_vgg = model.predict(x)
    x_vgg = x_vgg.reshape(1, 25088)
    return x_vgg

# List file names of the training datasets
folder = '/Users/gongbiao/Learning/Python/artificial_intelligence/week8/dataset/data_vgg/cats'
dirs = os.listdir(folder)

# Generate path for the images
img_path = []
for i in dirs:
    if os.path.splitext(i)[1] == '.jpg':
        img_path.append(i)
img_path = [folder + "//" + i for i in img_path]

# Preprocess multiple images
features1 = np.zeros([len(img_path), 25088])
for i in range(len(img_path)):
    feature_i = modelProcess(img_path[i], model_vgg)
    print('preprocessed:', img_path[i])
    features1[i] = feature_i
    
# ----------------
folder = '/Users/gongbiao/Learning/Python/artificial_intelligence/week8/dataset/data_vgg/dogs'
dirs = os.listdir(folder)
# Generate path for the dog images
img_path = []
for i in dirs:
    if os.path.splitext(i)[1] == '.jpg':
        img_path.append(i)
img_path = [folder + "//" + i for i in img_path]

# Preprocess multiple images
features2 = np.zeros([len(img_path), 25088])
for i in range(len(img_path)):
    feature_i = modelProcess(img_path[i], model_vgg)
    print('preprocessed:', img_path[i])
    features2[i] = feature_i

# Label the results
print(features1.shape, features2.shape)
y1 = np.zeros(300)
y2 = np.ones(300)

# Generate the training data
X = np.concatenate((features1, features2), axis = 0)
y = np.concatenate((y1, y2), axis = 0)
y = y.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Split the training and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 50)
print(X_train.shape, X_test.shape, X.shape)
```

```python
# Set up the mlp model
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(units = 10, activation='relu', input_dim = 25088))
model.add(Dense(units = 1, activation='sigmoid'))
model.summary()
```

```python
# Configure the model 
model.compile(optimizer= 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# Train the model
model.fit(X_train, y_train, epochs = 50)
```

```python
# Evaluate train accuracy score
from sklearn.metrics import accuracy_score
y_train_predict = model.predict(X_train)
y_train_predict = np.argmax(y_train_predict, axis = 1)
accuracy_train = accuracy_score(y_train, y_train_predict)
print(accuracy_train)
```

```python
# Evaluate test accuracy score
from sklearn.metrics import accuracy_score
y_test_predict = model.predict(X_test)
y_test_predict = np.argmax(y_test_predict, axis = 1)
accuracy_test= accuracy_score(y_test, y_test_predict)
print(accuracy_test)
```
