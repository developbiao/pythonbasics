```python
### Resize Image 
from PIL import Image
from torchvision import transforms

# Define resize operation
resize_img_oper = transforms.Resize((200, 200), interpolation=2)

# Original image
orig_img = Image.open('images/sunflower.png')
display(orig_img)

# Resize image
img = resize_img_oper(orig_img)
display(img)
```

```python
# Crop Image Operation
from PIL import Image
from torchvision import transforms

# Define operators
center_crop_oper = transforms.CenterCrop((60, 70))
random_crop_oper = transforms.RandomCrop((80, 80))
five_crop_oper = transforms.FiveCrop((60, 70))

# Original image
orig_img = Image.open('images/sunflower.png')
print('Original Image')
display(orig_img)

# Center crop
img1 = center_crop_oper(orig_img)
print('Center crop Image')
display(img1)

# Random Crop
img2 = random_crop_oper(orig_img)
print('Random crop Image')
display(img2)

# five crop Image
images = five_crop_oper(orig_img)
print('Five Crop Image')
for img in images:
    display(img)
```

```python
# Image flip
from PIL import Image
from torchvision import transforms

# Define flip operators
h_flip_oper = transforms.RandomHorizontalFlip(p=1)
v_flip_oper = transforms.RandomVerticalFlip(p=1)

# Original Image
orig_img = Image.open('images/sunflower.png')
display(orig_img)

# Horizontal Flip
print('Horizontal Flip')
img1 = h_flip_oper(orig_img)
display(img1)

# Vertial Flip
print('Vertial Flip')
img2 = v_flip_oper(orig_img)
display(img2)
```

```python
# Normalize
from PIL import Image
from torchvision import transforms

# Define normalize
norm_oper = transforms.Normalize((0.5, 0.5, 0.5, 0.5), (0.5, 0.5, 0.5, 0.5))

# Original image
orig_img = Image.open('images/sunflower.png')
display(orig_img)

# Convert image to tensor
img_tensor = transforms.ToTensor()(orig_img)


# Normalize 
tensor_norm = norm_oper(img_tensor)


# Tensor convert to image
print('Normalize image')
img_norm = transforms.ToPILImage()(tensor_norm)
display(img_norm)

```

```python
### Image compose
from PIL import Image
from torchvision import transforms

# Original image
orig_img = Image.open('images/sunflower.png')

# Composed operation
composed = transforms.Compose([transforms.Resize((200, 200)), transforms.RandomCrop(80)])

# Composed operator image
img = composed(orig_img)
display(img)
```

```python
# Mnist dataset transform
from torchvision import transforms
from torchvision import datasets

# Define a transoform
my_transform = transforms.Compose([transforms.ToTensor(),
                                   transforms.Normalize((0.5), (0.5))
                                  ])

# Read MNIST dataset and compose
mnist_dataset = datasets.MNIST(root='/mnt/d/workspace/model/data',
                               train = False,
                               transform = my_transform,
                               target_transform = None,
                               download = False
                              )

```
