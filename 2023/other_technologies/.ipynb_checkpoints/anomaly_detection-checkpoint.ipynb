### Anomaly Detection Check Practices
1.基于 anomaly_data.csv 数据，可视化数据分布情况、及对其高斯分布的概率密度函数。
2. 建立模型，实现异常数据点预测。
3. 可视化异常检查处理结果。
4. 修改概率分布阈值 EllipticEnvelope(contamination=0.1) 中的 contamination, 查看阈值改变对结果的影响。

```python
# Loda the data
import pandas as pd
import numpy as np
data = pd.read_csv('anomaly_data.csv')
data.head()
```

```python
# Visualize the data
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure(figsize = (10, 5))
plt.scatter(data.loc[:, 'x1'], data.loc[:, 'x2'])
plt.title('data')
plt.xlabel('x1')
plt.xlabel('x2')
plt.show()
```

```python
# Define x1 and x2
x1 = data.loc[:, 'x1']
x2 = data.loc[:, 'x2']

# Coding: utf-8
import matplotlib as mlp
fig2 = plt.figure(figsize=(20, 5))
plt.subplot(121)
plt.hist(x1, bins = 100)
plt.title('x1 distribution')
plt.xlabel('x1')
plt.ylabel('counts')

plt.subplot(122)
plt.hist(x2, bins = 100)
plt.title('x2 distribution')
plt.xlabel('x2')
plt.ylabel('counts')
plt.show()

```

```python
# Calculate the mean and sigma of x1 and x2
x1_mean = x1.mean()
x1_sigma = x1.std()
x2_mean = x2.mean()
x2_sigma = x2.std()
print(x1_mean, x1_sigma, x2_mean, x2_sigma)
```

```python
# Cacluate the guassian distribution p(x)
from scipy.stats import norm
x1_range = np.linspace(0, 20, 300)
x1_normal = norm.pdf(x1_range, x1_mean, x1_sigma)
x2_range = np.linspace(0, 20, 30)
x2_normal = norm.pdf(x2_range, x2_mean, x2_sigma)
```

```python
# Visualize the p(x)
fig2 = plt.figure(figsize=(20, 5))
plt.subplot(121)
plt.plot(x1_range, x1_normal)
plt.title('normal p(x1)')
plt.subplot(122)
plt.plot(x2_range, x2_normal)
plt.title('normal p(x2)')
plt.show()
```

```python
# Establish the model and predict
from sklearn.covariance import EllipticEnvelope
ad_model = EllipticEnvelope(contamination=0.03)
ad_model.fit(data)
```

```python
# Prediction
y_predict = ad_model.predict(data)
print(pd.value_counts(y_predict))
```

```python
# Visualize the result
fig4 = plt.figure(figsize = (10, 6))
original_data =plt.scatter(data.loc[:, 'x1'], data.loc[:, 'x2'], marker='x')
anomaly_data=plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',facecolor='none',edgecolor='red',s=150) 
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Auto find annomaly data')
plt.legend((original_data, anomaly_data), ('original data', 'annomaly data '))
plt.show()
```

```python
# Change contamination argument
ad_model = EllipticEnvelope(contamination = 0.02)
ad_model.fit(data)
y_predict = ad_model.predict(data)

# Visualize the result again
original_data =plt.scatter(data.loc[:, 'x1'], data.loc[:, 'x2'], marker='x')
anomaly_data=plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',facecolor='none',edgecolor='red',s=150) 
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Auto find annomaly data')
plt.legend((original_data, anomaly_data), ('original data', 'annomaly data '))
plt.show()
```
