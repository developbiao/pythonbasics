## Decision Tree Practices
1. 基于 iris_data.csv 数据，建立决策树模型，评估模型表现。
2. 可视化决策树结构。
3. 修改 min_samples_leaf 参数，对比模型结果。

```python
# Load the data
import pandas as pd
import numpy as py
data = pd.read_csv("iris_data.csv")
data.head()
```

```python
# Define the X and y
X = data.drop(['target', 'label'], axis = 1)
y = data.loc[:, 'label']
print(X.shape, y.shape)
```

```python
# Establish the decision tree model
from sklearn import tree
dc_tree = tree.DecisionTreeClassifier(criterion = 'entropy', min_samples_leaf = 5 )
dc_tree.fit(X, y)
```

```python
# Evaluate the model
y_predict = dc_tree.predict(X)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_predict)
print(accuracy)
```

```python
# Visualize the tree
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize = (20, 20))
tree.plot_tree(dc_tree, filled = 'True', feature_names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'], class_names=['setosa','versicolor','virginica'] )
```

```python
# Change min_samples_leaf
dc_tree = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=3)
dc_tree.fit(X,y)
fig = plt.figure(figsize=(8,8))
tree.plot_tree(dc_tree,filled='True',feature_names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'],class_names=['setosa','versicolor','virginica'])
```
